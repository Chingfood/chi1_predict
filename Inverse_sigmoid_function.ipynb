```python
N = 20
T_0 = 2
T_N_1 = 34
```

```python
import math
f_i = []
for i in range(N):
    f_i.append(1/(1+math.exp(4)) + i* (1/(1+math.exp(-4))-1/(1+math.exp(4)))/(N-1) )
```

```python
t_i = []
for fi in f_i:
    t_i.append((-math.log(1/fi -1)+4)* (T_N_1-T_0)/8 + T_0)
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()
ax.plot(range(N), f_i)

ax.grid(True, linestyle='-.')
ax.tick_params(labelcolor='r', labelsize='medium', width=3)

plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(range(N), t_i)


ax.tick_params(labelcolor='r', labelsize='medium', width=3)

plt.show()
```
